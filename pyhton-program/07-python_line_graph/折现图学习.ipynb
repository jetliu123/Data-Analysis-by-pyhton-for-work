```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei'] # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('兑券人数次数.xlsx')
df.head()
```

```python
# 学习绘制普通的曲线图
x = df['date']
y1 = df['兑券BA人数']
y2 = df['兑券数']
plt.style.use('ggplot')
plt.xticks(rotation = 90)
plt.xlabel('date')
plt.ylabel('兑换人数与券数')
plt.plot(x,y1)
plt.plot(x,y2)
plt.legend(['兑换人数','兑换券数'])
plt.show()
```

```python
## 一边画柱状图，一边画折线图
x = df['date']
y1 = df['兑券BA人数']
y2 = df['兑券数']
## plt.style.use('ggplot')
plt.xticks(rotation = 90)
plt.xlabel('date')
plt.ylabel('兑换人数与券数')
plt.bar(x,y1)
plt.plot(x,y2)
plt.legend(['兑换券数','兑换人数']) # 需要注意图标的位置，注意图表的格式是否一致
plt.show()
```

```python
## 一边画柱状图，一边画折线图 并且添加不同的颜色
## 一边画柱状图，一边画折线图
x = df['date']
y1 = df['兑券BA人数']
y2 = df['兑券数']
## plt.style.use('ggplot')
plt.xticks(rotation = 90)
plt.xlabel('date')
plt.ylabel('兑换人数与券数')
plt.bar(x,y1,color = 'green')
plt.plot(x,y2,color = 'blue',linestyle = 'dashed')
plt.legend(['兑换券数','兑换人数']) # 需要注意图标的位置，注意图表的格式是否一致
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

color = ['b','g','r','c','m','y','k','w']
line_style = ['-','--','-.',':']
dic1=[[0,1,2],[3,4,5]]
x=pd.DataFrame(dic1)
dic2=[[2,3,2],[3,4,3],[4,5,4],[5,6,5]]
y=pd.DataFrame(dic2)
# 循环输出所有"颜色"与"线型"
for i in range(2):
    for j in range(4):
        plt.plot(x.loc[i],y.loc[j],color[i*4+j]+line_style[j]) 
plt.show()
```
