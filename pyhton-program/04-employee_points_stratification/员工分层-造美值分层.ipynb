```python
import pandas as pd
import os 
```

```python
df = pd.read_excel('test0112.xlsx')
df.head()
```

```python
def classfied(x):
    if x ==0:
        result =  '一级区间'
    elif x >= 1 and x <= 150:
        result =  '二级区间'
    elif x >= 151 and x <= 1200:
         result = '三级区间'
    elif x >= 1201 and x <= 2700:
        result = '四级区间'
    elif x >= 2701 and x <= 3800:
        result =  '五级区间'
    else:
        result = '六级区间'
    return result
```

```python
df['区间'] = df['12月赚取造美值'].apply(classfied)
```
