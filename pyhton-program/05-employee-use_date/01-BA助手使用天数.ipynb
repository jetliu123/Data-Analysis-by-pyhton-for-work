```python
import pandas as pd
import os
```

```python
employee = pd.read_excel("1107.xlsx")
```

```python
ba = pd.read_excel("BA助手使用天数.xlsx")
```

```python
ba_use = pd.pivot_table(
                ba,
                index = 'ba_code',
                values = 'date',
                aggfunc = 'count'
)
```

```python
ba_use.head()
```

```python
ba_type = pd.merge(ba_use,
                  employee[['employeeNo','market','batype']],
                   left_on = "ba_code",
                   right_on = "employeeNo",
                   how = "left"
                  )
```

```python
ba_type.head()
```

```python
# 做一下筛选工作
ba_type = ba_type.loc[ba_type['batype'].isin(['Office','BA','EB','社群运营'])]
```

```python
ba_type.head()
ba_type.to_excel("测试问题在哪里.xlsx",index = False)
```

```python
# 分别处理batype 和 market 的统计情况
ba_use_day = pd.pivot_table(
                        ba_type,
                        index = ["batype","date"],
                        values = "employeeNo",
                        aggfunc = "count"
)
```

```python
ba_use_day
```

```python
market_use = pd.pivot_table(ba_type,
                            index = ['market','date'],
                            values = 'employeeNo',
                            aggfunc = 'count')
market_use
```
